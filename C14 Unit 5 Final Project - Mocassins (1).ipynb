```python
#index is 20-55
mocColors = df.columns.to_list()[20:55]

culture = df.loc[df["CULTURE"] == 5]

print(len(culture))

print(mocColors)
```

```python
plains = culture.iloc[:, 20:55]
plains['Total Colors'] = 0
plains['Name of Colors'] = 0
plains['Name of Colors'] = plains['Name of Colors'].astype("object")
plains['Total Colors'] = plains['Total Colors'].astype("object")

plains.head(5)

for index, row in plains.iterrows():
  colors = []
  for color in mocColors:
    if row.loc[color] == 1:
      colors.append(color)
      #.at is like .loc, but allows us to look at single values
      #this allows us to get singular column names and length
      #of the amount of colors and get them into a singular 
      #columns
      
  plains.at[index, 'Total Colors'] = len(colors)
  plains.at[index, 'Name of Colors'] = colors


plains.head(5)

#if the specific color column has a one, get column name 
```

```python
plainsMotifs = culture.iloc[:, 56:96]
plainsMotifs['Total Motifs'] = 0
plainsMotifs['Name of Motifs'] = 0
plainsMotifs['Name of Motifs'] = plainsMotifs['Name of Motifs'].astype("object")
plainsMotifs['Total Motifs'] = plainsMotifs['Total Motifs'].astype("object")

plainsMotifs.head(5)

for index, row in plainsMotifs.iterrows():
  motifs = []
  for motif in plainsMotifs:
    if row.loc[motif] == 1:
      motifs.append(motif)
  plainsMotifs.at[index, 'Total Motifs'] = len(motifs)
  plainsMotifs.at[index, 'Name of Motifs'] = motifs


plainsMotifs.head(5)
```
