```python
manateeDna = open("/content/drive/MyDrive/Colab Notebooks/Unit 4/4a_CYTB_DNA.txt").read()
dolphinDna = open("/content/drive/MyDrive/Colab Notebooks/Unit 4/4b_CYTB_DNA.txt").read()
cowDna = open("/content/drive/MyDrive/Colab Notebooks/Unit 4/4c_CYTB_DNA.txt").read()
```

```python
manRna = ""
dolRna = ""
cowRna = ""

for base in manateeDna:
  if base == "A": 
    manRna += "U"
  elif base == "T":
    manRna += "A"
  elif base == "G":
    manRna += "C"
  elif base == "C":
    manRna += "G"

for base in dolphinDna:
  if base == "A": 
    dolRna += "U"
  elif base == "T":
    dolRna += "A"
  elif base == "G":
    dolRna += "C"
  elif base == "C":
    dolRna += "G"

for base in cowDna:
  if base == "A": 
    cowRna += "U"
  elif base == "T":
    cowRna += "A"
  elif base == "G":
    cowRna += "C"
  elif base == "C":
    cowRna += "G"
```

```python
codonDict = {"UUU":"F","UUC":"F","UUA":"L","UUG":"L", "CUU":"L","CUC":"L",
             "CUA":"L","CUG":"L","AUU":"I","AUC":"I","AUA":"I","AUG":"M",
             "GUU":"V","GUC":"V","GUA":"V","GUG":"V","UCU":"S","UCC":"S",
             "UCA":"S","UCG":"S","CCU":"P","CCC":"P","CCA":"P","CCG":"P",
             "ACU":"T","ACC":"T","ACA":"T","ACG":"T","GCU":"A","GCC":"A",
             "GCA":"A","GCG":"A","UAU":"Y","UAC":"Y","UAA":"STOP","UAG":"STOP",
             "UGA":"STOP","CAU":"H","CAC":"H","CAA":"Q","CAG":"Q","AAU":"N",
             "AAC":"N","AAA":"K","AAG":"K","GAU":"D","GAC":"D","GAA":"E",
             "GAG":"E","UGU":"C","UGC":"C","UGG":"W","CGU":"R","CGC":"R",
             "CGA":"R","CGG":"R","AGU":"S","AGC":"S","AGA":"R","AGG":"R",
             "GGU":"G","GGC":"G","GGA":"G","GGG":"G"}
```

```python
proteinMan = ""
codonMan = ""
for i in manRna:
    if len(codonMan) == 3:
        proteinMan += codonDict[codonMan]
        codonMan = ""
        codonMan += i
    else:
        codonMan += i
print(len(proteinMan))

proteinDol = ""
codonDol = ""
for i in dolRna:
    if len(codonDol) == 3:
        proteinDol += codonDict[codonDol]
        codonDol = ""
        codonDol += i
    else:
        codonDol += i
print(len(proteinDol))

proteinCow = ""
codonCow = ""
for i in cowRna:
    if len(codonCow) == 3:
        proteinCow += codonDict[codonCow]
        codonCow = ""
        codonCow += i
    else:
        codonCow += i
print(len(proteinCow))
```

```python
numDifferencesManDol = 0

#For loop to iterate through BOTH lists
for c, i in enumerate(proteinDol):
  #Compare each sequence using an if statement
  if proteinMan[c] != i: #we found a difference!
    numDifferencesManDol += 1
  
print(numDifferencesManDol)

numDifferencesDolCow = 0

for c, i in enumerate(proteinCow):
  if proteinDol[c] != i: 
    numDifferencesDolCow += 1
  
print(numDifferencesDolCow)

numDifferencesManCow = 0

for c, i in enumerate(proteinCow):
  if proteinMan[c] != i:
    numDifferencesManCow += 1
  
print(numDifferencesManCow)

similarManDol = 100 - (numDifferencesManDol * 100/ len(proteinDol)) 
similarDolCow = 100 - (numDifferencesDolCow * 100/ len(proteinCow))
similarManCow = 100 - (numDifferencesManCow * 100/ len(proteinCow))
```

```python
#prints all similarity values
print(similarDolCow)
print(similarManCow)
print(similarManDol)
```
