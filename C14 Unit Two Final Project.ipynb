```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import pandas as pd
```

```python
csvFile = open("/content/drive/MyDrive/Colab Notebooks/sea_temp_anom_Sep_2011.csv", 'r') 
csvReader = csv.reader(csvFile, delimiter=',')
data_lists = list(csvReader)
#next(csvFile)

#new_dataframe = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/SeaTemp_SeaLevel_anomaly.csv", header=None,  skiprows=[0])
```

```python
height = len(data_lists)
width = len(data_lists[0])

#height = len(new_dataframe)

#width = len(new_dataframe[0])
```

```python
grid = np.empty([height, width, 3], dtype=np.uint8)
```

```python
print(data_lists)
```

```python
all_values = [] # We will fill in this empty list with all the data values as we go
#storing color values in all values 

#storing color values in grid

for r, row in enumerate(data_lists):
    for c, value in enumerate(row):
        if float(value) < 99998:
            all_values.append(float(value)) 
        if float(value) <= -5:
            grid[r, c] = [46, 46, 255]
        elif float(value) <= -4:
            grid[r, c] = [97, 97, 255]
        elif float(value) <= -3:
            grid[r, c] = [145, 169, 255]
        elif float(value) <= -2:
            grid[r, c] = [199, 211, 255]
        elif float(value) <= -1:
            grid[r, c] = [230, 230, 255]
        elif float(value) <= 0:
            grid[r, c] = [245, 245, 245]
        elif float(value) <= 1:
            grid[r, c] = [255, 181, 181]
        elif float(value) <= 2:
            grid[r, c] = [255, 150, 150]
        elif float(value) <= 3:
            grid[r, c] = [255, 117, 117]
        elif float(value) <= 4:
            grid[r, c] = [255, 79, 79]
        elif float(value) <= 5:
            grid[r, c] = [255, 43, 43]
        else:
            grid[r, c] = [0, 0, 0]
```

```python
print(min(all_values))

print(max(all_values))
```

```python
plt.title("Sea Temperature Anomalies In September of 2011")
plt.imshow(grid)
plt.axis('off')
plt.savefig("Sea_temperature_2002.png",dpi = 1000)
plt.show()
```
